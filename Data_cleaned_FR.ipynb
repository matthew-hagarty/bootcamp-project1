```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
```

```python
france_data_json = pd.read_json('Resources/FR_category_id.json')
fr_df = pd.read_csv('Resources/FR_youtube_trending_data.csv')
```

```python
# jsons
with open('Resources/FR_category_id.json') as fr_file:
    fr_data = json.load(fr_file)
#print(fr_data)
categoryNum = []
for category in fr_data['items']:
    categoryNum.append(category['id'])
categoryNames = []
for category in fr_data['items']:
    categoryNames.append(category['snippet']['title'])
categoryDict = {categoryNum[i]: categoryNames[i] for i in range(len(categoryNum))}
categoryDict['29'] = 'Nonprofits & Activism'
categoryDict
```

```python
fr_df['categoryName'] = [categoryDict[str(x)] for x in fr_df['categoryId']]
fr_df
```

```python
fr_df = fr_df.drop(fr_df.loc[fr_df['view_count'] == 0].index)
fr_df = fr_df.drop(fr_df.loc[fr_df['comments_disabled'] == True].index)
fr_df = fr_df.drop(fr_df.loc[fr_df['ratings_disabled'] == True].index)
fr_df
```

```python
# convert the date published column to a date type
fr_df['publishedAt'] = pd.to_datetime(fr_df['publishedAt'])
# add a year, month, and day of week column
fr_df['month'] = fr_df['publishedAt'].dt.month_name()
fr_df['weekday'] = fr_df['publishedAt'].dt.day_name()
fr_df['year'] = fr_df['publishedAt'].dt.year
fr_df = fr_df.drop(fr_df.loc[fr_df['year'] == 2023].index)
fr_df
```

```python
fr_gb = fr_df.groupby(['year', 'categoryName'])['view_count'].mean().reset_index()
```

```python
# Pivot the data to have categories as columns and years as rows
fr_pivot = fr_gb.pivot(index='year', columns='categoryName', values='view_count')
# Plot the line graph
fig, ax = plt.subplots(figsize=(12, 12))
fr_pivot.plot(kind='line', ax=ax, marker='o')
# Move the legends outside the graph
ax.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
# Set labels and title
years = np.arange(2020, 2023)
ax.set_xticks(years)
ax.set_xlabel('Year')
ax.set_ylabel('Average Views')
ax.set_title('Average Views by Category and Year')
```

```python
# calculate average engagement by category
viewsByCategory = fr_df.groupby(['categoryName'])['view_count'].mean()
likesByCategory = fr_df.groupby(['categoryName'])['likes'].mean()
commentsByCategory = fr_df.groupby(['categoryName'])['comment_count'].mean()
```

```python
# Sort the data by 'view_count' in descending order
viewsByCategory_sorted = viewsByCategory.sort_values(ascending=False)
# Sort the data by 'likes' in descending order
likesByCategory_sorted = likesByCategory.sort_values(ascending=False)
# Sort the data by 'comment_count' in descending order
commentsByCategory_sorted = commentsByCategory.sort_values(ascending=False)
# Plot the data from the previous cell
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.suptitle("Average Total Engagement by Category")
fig.set_figheight(5)
fig.set_figwidth(18)
viewsByCategory_sorted.plot(kind='bar', title='Avg. Views by Category', ylabel='Avg. View Count (millions)', xlabel='Category', ax=ax1)
likesByCategory_sorted.plot(kind='bar', title='Avg. Likes by Category', ylabel='Avg. Likes', xlabel='Category', ax=ax2, color='red')
commentsByCategory_sorted.plot(kind='bar', title='Avg. Comments by Category', ylabel='Avg. Comment Count', xlabel='Category', ax=ax3, color='green')
plt.show()
```
