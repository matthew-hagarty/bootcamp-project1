```python
# jsons
with open('Resources/FR_category_id.json') as fr_file:
    fr_data = json.load(fr_file)
#print(fr_data)
categoryNum = []
for category in fr_data['items']:
    categoryNum.append(category['id'])
categoryNames = []
for category in fr_data['items']:
    categoryNames.append(category['snippet']['title'])
categoryDict = {categoryNum[i]: categoryNames[i] for i in range(len(categoryNum))}
categoryDict['29'] = 'Nonprofits & Activism'
categoryDict
```

```python
fr_df = france_data_csv
fr_df.head()
```

```python
fr_df['categoryName'] = [categoryDict[str(x)] for x in fr_df['categoryId']]
    
fr_df.head()
```

```python
fr_df.describe()
```

```python
fr_df['categoryId'].value_counts()
```

```python
fr_df.columns
```

```python
fr_df['publishedAt'] = pd.to_datetime(fr_df['publishedAt'])
fr_df['publishDate'] = fr_df['publishedAt'].dt.date
fr_df['publishTime'] = fr_df['publishedAt'].dt.time
fr_df.head()
```

```python
fr_df['trending_date'] = pd.to_datetime(fr_df['trending_date']).dt.date
fr_df.head()
```

```python
fr_df = fr_df.drop(fr_df.loc[fr_df['view_count'] == 0].index)
fr_df
```

```python
fr_df = fr_df.drop(fr_df.loc[fr_df['ratings_disabled'] == True].index)
fr_df = fr_df.drop(fr_df.loc[(fr_df['trending_date'] < pd.Timestamp('2021-12-15')) & (fr_df['dislikes'] == 0)].index)
fr_df
```

```python
fr_df['comment_pct'] = fr_df['comment_count'] / fr_df['view_count'] * 100
fr_df['likes_pct'] = fr_df['likes'] / fr_df['view_count'] * 100
fr_df['like_ratio'] = fr_df['likes'] / fr_df['dislikes']
fr_df.head()
```

```python
fr_df_groupby = fr_df.groupby(['video_id'])['trending_date'].max()
fr_df_groupby.rename_axis('max_trending_date')
fr_df_groupby
```

```python
fr_df2 = pd.merge(fr_df, fr_df_groupby, how = 'left', on = 'video_id')
fr_df2
```

```python
fr_df_groupby2 = fr_df2.groupby(['video_id'])
fr_min = fr_df_groupby2['trending_date_x'].min()
fr_cleaned = pd.merge(fr_df2, fr_min, how = 'inner', on = ['video_id', 'trending_date_x'])
fr_cleaned
```

```python
fr_cleaned.columns
```

```python
fr_final = fr_cleaned[['video_id', 'title', 'categoryName', 'publishDate', 'publishTime', 'view_count', 'likes', 'dislikes', 'comment_count', 'likes_pct', 'like_ratio', 'comment_pct', 'trending_date_x', 'trending_date_y']].reindex()
```
